#### while문
- 반복횟수가 명확하지 않을 때

```python
# 파이썬 최고! 13번 출력해보기
num = 0
while num < 13:
    num += 1
    print(f'{num}. 파이썬 최고!')
```

```python
while True:
    print('무한루프')
    break
```

```python
while True:
    num1 = int(input('첫 번째 정수 입력 >> '))
    num2 = int(input('두 번째 정수 입력 >> '))
    
    if num1 or num2 != 0:
        print('두 정수의 합 : ', num1 + num2)
    else:
        print('프로그램이 종료되었습니다.')
        break
```

```python
C_kg = int(input('현재 몸무게 : '))
T_kg = int(input('목표 몸무게 : '))
week = 1
while C_kg >= T_kg:
    M_kg = int(input(f'{week}주차 감량 몸무게 : '))
    week += 1
    C_kg -= M_kg
print(f'{C_kg}kg 달성! 축하합니다!!')
```

```python
ran = random.randint(1,50)

while True:
    num = int(input('숫자를 입력하세요 >> '))
    if ran == num:
        print('점수를 맞추셨습니다.')
        break
    elif ran > num:
        print(f'{num}보다 큰 수 입니다.')
    elif ran < num:
        print(f'{num}보다 작은 수 입니다.')
```

##### for : range()
- 필요한 만큼의 숫자를 만들어내는 유용한 기능
- range(시작할 숫자, 종료할 숫자, 증감량)

```python
lunch = ['샌드위치', '백반', '돈까스']
```

```python
for i in range(0, len(lunch), 1): # 시작 인덱스가 0과 같거나, 증감값(defalut : 1) --> 생략 가능
    print(i, lunch[i])
```

```python
score_list = [90, 45, 70, 60, 55]
for num in range(0, len(score_list), 1): # 0 ~ 리스트의 길이만큼 숫자 반환
    
    if score_list[num] >= 60: # i : 인덱스 번호, score_list[i] ==> 데이터에 접근
        print(f'{num + 1}번 학생은 합격입니다.')
    else:
        print(f'{num + 1}번 학생은 불합격입니다.')
```

```python
for i in range(1, 25, 1):
    print(i, end = ' ')
```

- min, max 함수
- 선언한 자료구조 (리스트, 튜플)에서 가장 큰 값과 가장 작은 값을 찾아주는 함수


```python
list2 = [4, 5, 6, 3, 2, 1, 66, 8, 9, 34]
min(list2)
```

```python
# 가장 큰 수 구하기
max(list2)


max_num = list2[0]

for i in list2:
    if max_num < i:
        max_num = i
print(max_num)
```

```python
# 1 ~ 100 사이의 숫자 중에서 3의 배수의 합을 구하기
plus = 0
for i in range(0, 101, 3):
    plus += i
print(plus)
```
