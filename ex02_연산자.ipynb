##### 숫자, 문자열을 더했을 때

```python
num1 = 10 # int 정수 자료형
str1 = "7" # str 문자열 자료형

# print(num1+str1) --> 오류
# 형변환(int) --> (int)X --> int(대상)
# 숫자가 연산되도록 형변환
print(num1 + int(str1))
# 문자가 연산되도록 형변환
print(str(num1) + str1)
```

#### 비교연산자

```python
# True, False로 결과가 출력되는 연산자
# >, >=, <, <=, ==(같다), !=(다르다)
```

```python
a = 3
b = 7
print(a > b) # False
print(a <= b) # True
print(a == b) # False
print(a != b) # True
```

#### 논리연산자

```python
# not : 논리를 뒤집어버리는 연산자 True -> False, False -> True
# and, or
```

```python
# not
a, b = 5, 10
not a < b # True -> False
```

```python
# and, or
# and : 두 개의 논리를 연결하여 확인, 두 논리가 모두 참일 경우 -> True
# or : 한 개의 논리만 참일 경우에도 True

3 > 5 and 10 == 10 # false
3 > 5 or 10 == 10 # True
```

 #### 삼항연산자
 - (True일 때 실행문장) if (조건식) else (False일 때 실행문장)

```python
# 60점 이상이면 합격, 미만이면 불합격 출력해보기
```

```python
score = 70
print('합격' if score >= 60 else '불합격')
```

```python
a = int(input('정수입력 >> '))
b = int(input('정수입력 >> '))

f'a : {a}' if a > b else f' b: {b}' # b가 참일 경우 (크다)

f'a : {a}' if a < b else f' b: {b}' # a가 참일 경우 (작다)
```
